# Import FASTA Files for Training

```python
refPathTrain = os.path.join('CMPBIO290-2016_project/','fasta_files/','fasta_ref_4-eQTL-causal_SNPs_sample_v2_TRAINING.fa')
altPathTrain = os.path.join('CMPBIO290-2016_project/','fasta_files/','fasta_alt_4-eQTL-causal_SNPs_sample_v2_TRAINING.fa')

pdSeqTrain = pd.DataFrame(columns = ['ID', 'Ref', 'Alt'])

pdSeqTrain['ID'] = pdTrain['ID']

refFile = open(refPathTrain)
altFile = open(altPathTrain)

currSeq = ''
currLine = ''

for line in refFile:
    if line[0] == '>':
        currSeq = str.split(str.split(line, '\n')[0], '>')[1]
    else:
        pdSeqTrain.loc[pdSeqTrain['ID'] == currSeq, 'Ref'] = str.split(line, '\n')[0]
        
for line in altFile:
    if line[0] == '>':
        currSeq = str.split(str.split(line, '\n')[0], '>')[1]
    else:
        pdSeqTrain.loc[pdSeqTrain['ID'] == currSeq, 'Alt'] = str.split(line, '\n')[0]

refFile.close()
altFile.close()
```

# Filter the Training DataSet into Positive and Negative Sequences (Full length)

```python
posSeq = dict()
negSeq = dict() # these are the variants of ref and alt where one hits as a reg hit and one does not
for entry in pdTrain.loc[pdTrain['Regulatory_Hit'] == 1, 'ID']:
    if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.A.logP'] > 6.6:
        if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 6.6:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
        else:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            negSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
    elif pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 6.6:
        negSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
        posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
```

```python
negSeq2 = dict() # these are the seq with Regulatory Hit = 0 (both Ref and Alt) (This set is more generous)
for entry in pdTrain.loc[pdTrain['Regulatory_Hit'] == 0, 'ID']:
    if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.A.logP'] > 6.6:
        print('Bad Theory 1')
    elif pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 6.6:
        print('Bad Theory 2')
    else:
        negSeq2[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
        negSeq2[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
```

```python
print len(posSeq)
print len(negSeq)
print len(negSeq2)
```

# Write posSeq, negSeq, negSeq2 to FASTA files

```python
# Create new file object outputFile
outputFilePos = open('posSeq.fa','w')
outputFileNeg = open('negSeq.fa','w')
outputFileNeg2 = open('negSeq2.fa','w')

# Based on each sequence in posSeq, negSeq, negSeq2, write an entry to the associated output file 
for key in posSeq:
    outputFilePos.write('>' + key + '\n')
    outputFilePos.write(posSeq[key] + '\n')

for key in negSeq:
    outputFileNeg.write('>' + key + '\n')
    outputFileNeg.write(negSeq[key] + '\n')
    
for key in negSeq2:
    outputFileNeg2.write('>' + key + '\n')
    outputFileNeg2.write(negSeq2[key] + '\n')   
    
# Close the output file
outputFilePos.close()
outputFileNeg.close()
outputFileNeg2.close()
```

# Make .fa Files with Higher Stringency

```python
posSeq = dict()
negSeq = dict() # these are the variants of ref and alt where one hits as a reg hit and one does not
for entry in pdTrain.loc[pdTrain['Regulatory_Hit'] == 1, 'ID']:
    if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.A.logP'] > 10:
        if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 10:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
        else:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            negSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
    elif pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 10:
        negSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
        posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
        
# Create new file object outputFile
outputFilePos = open('posSeq_10.fa','w')
outputFileNeg = open('negSeq_10.fa','w')

# Based on each sequence in posSeq, negSeq, negSeq2, write an entry to the associated output file 
for key in posSeq:
    outputFilePos.write('>' + key + '\n')
    outputFilePos.write(posSeq[key] + '\n')

for key in negSeq:
    outputFileNeg.write('>' + key + '\n')
    outputFileNeg.write(negSeq[key] + '\n')
     
    
# Close the output file
outputFilePos.close()
outputFileNeg.close()
```

# Make .fa Files with Random Sampling for Positives


```python
posSeq = dict()
negSeq = dict() # these are the variants of ref and alt where one hits as a reg hit and one does not
for entry in pdTrain.loc[pdTrain['Regulatory_Hit'] == 1, 'ID']:
    if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.A.logP'] > 100:
        if pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 100:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
        else:
            posSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
            negSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
    elif pdTrain.loc[pdTrain[pdTrain['ID'] == entry].index[0], 'C.B.logP'] > 100:
        negSeq[entry + '_Ref'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Ref']
        posSeq[entry + '_Alt'] = pdSeqTrain.loc[pdSeqTrain[pdSeqTrain['ID'] == entry].index[0], 'Alt']
        
# Create new file object outputFile
outputFileSample = open('posSeq_100_5a.fa','w')
outputFileScore = open('posSeq_100_5b.fa','w')

# Based on each sequence in posSeq, negSeq, negSeq2, write an entry to the associated output file 
numHits = len(posSeq)
print(numHits)
sampleHits = int(floor(0.8*numHits))

posKeys = posSeq.keys()
shuffle(posKeys)
sampleKeys = posKeys[0:sampleHits]
scoreKeys = posKeys[sampleHits:]

for key in sampleKeys:
    outputFileSample.write('>' + key + '\n')
    outputFileSample.write(posSeq[key] + '\n')

for key in scoreKeys:
    outputFileScore.write('>' + key + '\n')
    outputFileScore.write(posSeq[key] + '\n')   
    
# Close the output file
outputFileSample.close()
outputFileScore.close()
```
